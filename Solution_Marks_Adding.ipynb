```python
import csv

file = open('marks.csv', 'r')
csvreader = csv.reader(file)
header = next(csvreader)

totalmarks = {}
for row in csvreader:
  x = row[0]
  if x in totalmarks:
    totalmarks[x] = totalmarks[x] + int(row[1])
  else:
    totalmarks[x] = int(row[1])

print(totalmarks)
```

Method 2 using Pandas in Python with iterrows()

```python
import pandas as pd
import numpy as np
df = pd.read_csv('marks.csv')

totalmarks = {}

for index, row in df.iterrows():
  x = row[df.columns[0]]
  if x in totalmarks:
    totalmarks[x] = totalmarks[x] + int(row[df.columns[1]])
  else:
    totalmarks[x] = int(row[df.columns[1]])

# for i in sorted(totalmarks):
#   print(i, totalmarks[i])
print(totalmarks)

```

Method 3 using Pandas in Python with DataFrame.iloc[row_range, column_list]

```python
import pandas as pd
import numpy as np
df = pd.read_csv('marks.csv')

totalmarks = {}

for i in range(len(df)):
  x = df.iloc[i,0]
  if x in totalmarks:
    totalmarks[x] = totalmarks[x] + int(df.iloc[i, 1])
  else:
    totalmarks[x] = int(df.iloc[i, 1])

for i in sorted(totalmarks):
  print(i, totalmarks[i])
```
